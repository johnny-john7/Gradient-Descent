# what is gradient descent

Gradient descent is an optimization algorithm used to minimize the loss function in machine learning and deep learning. It is an iterative method that adjusts the parameters of a model to find the values that minimize the difference between the model's predictions and the actual output.


### How Does it Work?

Start with a guess:   Make an initial guess at the solution.
Check how good it is: Calculate how good or bad the guess is by comparing it to the correct answer.
Make it better:       Make small changes to the guess to make it better.
Repeat:               Keep making small changes and checking how good the solution is until it's as good as it can be.

# Here's a simple problem and solution using gradient descent Problem:

### Suppose we want to find the best fit line for the following data points:

        x	y
        1	2
        2	3
        3	5
        4	7
        5	11

The goal is to find the values of m and b in the equation y = mx + b that minimize the difference between the predicted y values and the actual y values.

Loss Function:

We'll use the mean squared error (MSE) as the loss function, which is defined as:

L(m, b) = (1/n) * Σ(y_i - (mx_i + b))^2

where n is the number of data points, y_i is the actual y value, x_i is the actual x value, and m and b are the parameters to be optimized.

Gradient Descent Solution:

We'll use gradient descent to minimize the loss function L(m, b). The gradient of the loss function with respect to m and b is:

∂L/∂m = -(2/n) * Σx_i * (y_i - (mx_i + b))

∂L/∂b = -(2/n) * Σ(y_i - (mx_i + b))

The update rules for m and b are:

m_new = m_old - α * ∂L/∂m

b_new = b_old - α * ∂L/∂b

where α is the learning rate.

```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.08

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,b_curr,cost, i))

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```
